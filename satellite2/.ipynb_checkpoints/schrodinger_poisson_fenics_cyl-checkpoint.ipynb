```python
from dolfin import (FiniteElement,File, TestFunctions,Function,Point,MixedElement,
                    FunctionSpace,split,Expression,Constant, triangle, solve, plot, dx,
                    dot, grad, DirichletBC, project, near, ds, MeshFunction)
from dolfin import *
from mshr import Rectangle,generate_mesh, Circle
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import *
import plots as pts
```

```python
rhomin = 0.01
Rf = 100.
###########################     MESH         ##################################
circulo = Circle(Point(0.0,0.0), Rf) - Rectangle(Point(-Rf,-Rf),Point(0.,Rf)) - Rectangle(Point(0.,-Rf), Point(Rf,0.)) - Rectangle(Point(rhomin,0.),Point(0.,Rf))
mesh = generate_mesh(circulo, 164)#
#plot(mesh)
```

```python
boundary_markers = MeshFunction("size_t", mesh, mesh.topology().dim()-1, 0)

class leftwall(SubDomain):
    tol = 1E-14
    def inside(self, x, on_boundary):
        return on_boundary and near(x[0], rhomin, tol)
tol = 1E-14    
class bottomwall(SubDomain):
    tol = 1E-14
    def inside(self, x, on_boundary):
        return on_boundary and near(x[1], 0, tol)

left = leftwall()
bottom = bottomwall()
```

```python
left.mark(boundary_markers, 1)
bottom.mark(boundary_markers, 2)
```

```python
ds = Measure('ds', domain=mesh, subdomain_data=boundary_markers)
```

```python
P1 = FiniteElement('P', triangle, 1)
V = FunctionSpace(mesh, MixedElement([P1, P1]))
V0 = FunctionSpace(mesh, 'P', 3)
############    Define test functions      #####################################
v_1, v_2= TestFunctions(V)
############     Define functions for SF and potential      #########
u = Function(V)
#############         Split system functions to access components      ########
psi, Phi = split(u)
###############          Define expressions used in variational forms
E= -0.034465
fx = Expression('a*x[0] + b', degree=1, a=0., b= 1.)
k = Constant(E)
g1 = Constant(0.)
g2 = Constant(0.)
g3 = Constant(0.)
##############          Define variational problem     #########################
F = fx*dot(grad(psi), grad(v_1))*dx + fx*dot(grad(Phi), grad(v_2))*dx \
+ fx*Phi*Phi*v_1*dx + fx*psi*Phi*v_2*dx + fx*k*Phi*v_2*dx - g1*v_1*ds(2) - g2*v_1*ds(1) - g3*v_2*ds(1)

#############           Define boundaries             #########################
tol = 1E-10
def leftwall(x):
    return near(x[0], rhomin, tol)

def buttonwall(x):
    return near(x[1], 0, tol)

#def circlebord(x):
##    return abs(np.sqrt(x[0]*x[0] + x[1]*x[1]) - 10.) < tol 
#    return near(np.sqrt(x[0]*x[0] + x[1]*x[1]), 10. ,tol) 

def circlebord(x, on_boundary):
    if on_boundary:
        if near(x[0], rhomin, tol) or near(x[1],0, tol):
            return False
        else:
            return True
    else:
        return False
```

```python
##############          boundary conditions         ########################
bc = DirichletBC(V.sub(0), Constant(-0.1) , circlebord) # on psi
bc2 = DirichletBC(V.sub(1), Constant(0.001) , circlebord)  # on Phi
bc3 = DirichletBC(V.sub(1), Constant(0.) , buttonwall)  # on Phi
```

```python
solve(F == 0, u, [bc,bc2, bc3])    # Solve variational problem 
```

```python
bc = DirichletBC(V.sub(0), Expression('1/sqrt(pow(x[0],2) + pow(x[1],2))', degree=2) , circlebord) # on psi
#bc = DirichletBC(V.sub(0), Constant(0.) , circlebord) # on psi
bc1 = DirichletBC(V.sub(1), Constant(0.) , circlebord)  # on Phi
bc2 = DirichletBC(V.sub(1), Constant(0.) , buttonwall)  # on Phi
```

```python
solve(F == 0, u, [bc,bc1,bc2])    # Solve variational problem 
```

```python
# Save solution to file (VTK)
_u_1, _u_2 = u.split()
```

```python
plt.figure()
plot(_u_1, title="Potential")
plt.xlabel('$\\mu \\rho$')
plt.ylabel('$\\mu z$')
#plt.figure()
#plot(grad(_u_1), title="Potential gradient")
plt.figure()
plot(_u_2, title="Scalar field $\\Phi$")
plt.xlabel('$\\mu\\rho$')
plt.ylabel('$\\mu z$')
plt.show()
```

```python
u.set_allow_extrapolation(True)
tol = 0.001 # avoid hitting points outside the domain

z = np.linspace(0 + tol, 100. - tol, 101)
rho = np.linspace(0 + tol, 100. - tol, 101)
points = [(5., z_) for z_ in z]
Philine = np.array([u[1](point) for point in points])
psiline = np.array([u[0](point) for point in points])

Phi = []
for i in range(0,np.shape(rho)[0]):    
    points = [(rho[i], z_) for z_ in z] # 2D points
    Phi.append([u[1](point) for point in points])
Phi = np.array(Phi)    

psi = []
for i in range(0,np.shape(rho)[0]):    
    points = [(rho[i], z_) for z_ in z] # 2D points
    psi.append([u[0](point) for point in points])
psi = np.array(psi)

pts.parametricplot(z,Philine,'$z$','$\\Phi (5,z) $','title','nom_archivo', save = False)

pts.parametricplot(z,psiline,'$z$','$\\psi (5,z) $','title','nom_archivo', save = False)

Rho, Z = np.meshgrid(rho, z)

pts.densityplot(Rho,Z,Phi,r'$\mu \rho$',r'$\mu z$',r'$\Phi(\rho,z)$','','')
pts.densityplot(Rho,-Z,-Phi,r'$\mu \rho$',r'$\mu z$',r'$\Phi(\rho,z)$','','')
pts.densityplot(Rho,Z,psi,r'$\mu \rho$',r'$\mu z$',r'$\psi(\rho,z)$','','')
pts.densityplot(Rho,-Z,psi,r'$\mu \rho$',r'$\mu z$',r'$\psi(\rho,z)$','','')
pts.plotfunc3d(Rho,Z,Phi,r'$\mu \rho$',r'$\mu z$',r'$\Phi(\rho,z)$','','')
pts.plotfunc3d(Rho,Z,psi,r'$\mu \rho$',r'$\mu z$',r'$\psi(\rho,z)$','','')
```
