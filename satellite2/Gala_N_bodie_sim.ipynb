```python
>>> import astropy.units as u
>>> import numpy as np
>>> import gala.potential as gp
>>> import gala.dynamics as gd
>>> from gala.dynamics.nbody import DirectNBody
>>> from gala.units import galactic, UnitSystem
import matplotlib.pyplot as plt
%matplotlib inline
```

Let’s create two initial conditions by specifying the position and velocity of two particles, then combine them into a single PhaseSpacePosition object

```python
>>> w0_1 = gd.PhaseSpacePosition(pos=[0, 0, 0] * u.pc,
...                              vel=[0, 1.5, 0] * u.km/u.s)
>>> w0_2 = gd.PhaseSpacePosition(pos=w0_1.xyz + [100., 0, 0] * u.pc,
...                              vel=w0_1.v_xyz + [0, 5, 0] * u.km/u.s)
>>> w0 = gd.combine((w0_1, w0_2))
>>> w0.shape
```

We’ll then treat particle 1 as a massive object by sourcing a HernquistPotential at the location of the particle, and particle 2 as a test particle: To treat some particles as test particles, you can pass None or a NullPotential instance for the corresponding particle potential:

```python
pot1 = gp.HernquistPotential(m=1e7*u.Msun, c=0.5*u.kpc, units=galactic)
particle_pot = [pot1, None]
```

Let’s now create an N-body instance and try integrating the orbits of the two particles. Here, there is no external potential, so particle 1 (the massive particle) will move off in a straight line. We’ve created the initial conditions for particle 2 so that it will remain bound to the potential sourced by particle 1, and so will orbit it as it moves.

```python
>>> nbody = DirectNBody(w0, particle_pot)
>>> orbits = nbody.integrate_orbit(dt=1e-2*u.Myr, t1=0, t2=1*u.Gyr)
>>> fig, ax = plt.subplots(1, 1, figsize=(5, 5)) 
>>> _ = orbits[:, 0].plot(['x', 'y'], axes=[ax]) 
>>> _ = orbits[:, 1].plot(['x', 'y'], axes=[ax]) 
```

```python
figs = orbits[:, 0].plot(marker=',', linestyle='none')
```

```python
figs = orbits[:, 1].plot(marker=',', linestyle='none')
```

```python
w0 = orbits.w()[0]
w1 = orbits.w()[1]
np.shape(w1)
```

Example: N-body integration with a background potential

```python
>>> external_pot = gp.MilkyWayPotential()
>>> w0_1 = gd.PhaseSpacePosition(pos=[10, 0, 0] * u.kpc,
...                              vel=[0, 200, 0] * u.km/u.s)
>>> w0_2 = gd.PhaseSpacePosition(pos=w0_1.xyz + [10., 0, 0] * u.pc,
...                              vel=w0_1.v_xyz + [0, 5, 0] * u.km/u.s)
>>> w0 = gd.combine((w0_1, w0_2))
>>> pot1 = gp.HernquistPotential(m=1e7*u.Msun, c=0.5*u.kpc, units=galactic)
>>> particle_pot = [pot1, None]
>>> nbody = DirectNBody(w0, particle_pot, external_potential=external_pot)
>>> orbits = nbody.integrate_orbit(dt=1e-2*u.Myr, t1=0, t2=1*u.Gyr)
>>> fig, ax = plt.subplots(1, 1, figsize=(5, 5)) 
>>> _ = orbits[:, 0].plot(['x', 'y'], axes=[ax]) 
>>> _ = orbits[:, 1].plot(['x', 'y'], axes=[ax])
```

```python
fig= plt.figure()
orbits[:, 0].plot(marker=',', linestyle='none')
orbits[:, 1].plot(marker=',', linestyle='none')
plt.show()
```
